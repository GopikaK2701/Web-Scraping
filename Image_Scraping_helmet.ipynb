```python
from bs4 import *
import requests as rq
import os

r1 = rq.get("https://www.freeimages.com/search/construction-hat")
r2 = rq.get("https://www.freeimages.com/search/construction-hat/2")
r3 = rq.get("https://www.freeimages.com/search/construction-hat/3")
r4 = rq.get("https://www.freeimages.com/search/construction-hat/4")
r5 = rq.get("https://www.freeimages.com/search/construction-hat/5")
r6 = rq.get("https://www.freeimages.com/search/construction-hat/6")
r7 = rq.get("https://www.freeimages.com/search/construction-hat/7")
r8 = rq.get("https://www.freeimages.com/search/construction-hat/8")
r9 = rq.get("https://www.freeimages.com/search/construction-hat/9")
r10 = rq.get("https://www.freeimages.com/search/construction-hat/10")
            
soup1 = BeautifulSoup(r1.text, "html.parser")
soup2 = BeautifulSoup(r2.text, "html.parser")
soup3 = BeautifulSoup(r3.text, "html.parser")
soup4 = BeautifulSoup(r4.text, "html.parser")
soup5 = BeautifulSoup(r5.text, "html.parser")
soup6 = BeautifulSoup(r6.text, "html.parser")
soup7 = BeautifulSoup(r7.text, "html.parser")
soup8 = BeautifulSoup(r8.text, "html.parser")
soup9 = BeautifulSoup(r9.text, "html.parser")
soup10 = BeautifulSoup(r10.text, "html.parser")
links=[]
x1 = soup1.select('img[src^="https://images.freeimages.com/"]')
x2 = soup2.select('img[src^="https://images.freeimages.com/"]')
x3 = soup3.select('img[src^="https://images.freeimages.com/"]')
x4 = soup4.select('img[src^="https://images.freeimages.com/"]')
x5 = soup5.select('img[src^="https://images.freeimages.com/"]')
x6 = soup6.select('img[src^="https://images.freeimages.com/"]')
x7 = soup7.select('img[src^="https://images.freeimages.com/"]')
x8 = soup8.select('img[src^="https://images.freeimages.com/"]')
x9 = soup9.select('img[src^="https://images.freeimages.com/"]')
x10 = soup10.select('img[src^="https://images.freeimages.com/"]')

for img in x1:
    links.append(img['src'])
for img in x2:
    links.append(img['src'])
for img in x3:
    links.append(img['src'])
for img in x4:
    links.append(img['src'])
for img in x5:
    links.append(img['src'])
for img in x6:
    links.append(img['src'])
for img in x7:
    links.append(img['src'])
for img in x8:
    links.append(img['src'])
for img in x9:
    links.append(img['src'])
for img in x10:
    links.append(img['src'])

os.makedirs("basedata/training/helmet")
i=1
for index, img_link in enumerate(links):
    if i<=800:
        img_data = rq.get(img_link).content
        with open('basedata/training/helmet//'+str(index+1)+'.jpg','wb+') as f:
            f.write(img_data)
        i+=1
    else:
        f.close()
        break  
```
